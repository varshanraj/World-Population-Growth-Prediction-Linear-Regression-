```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/World Population Growth.csv', thousands=',')
```

```python
#df['Yearly Growth %'] = df['Yearly Growth %'].str.rstrip('%').astype('float') / 100

x = df['Year']
y = df['Population']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=21)
x_train = np.array(x_train).reshape(-1,1)
y_train = np.array(y_train).reshape(-1,1)
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(np.array(x_test).reshape(-1,1))

plt.plot(x_test, y_test, color='blue', label='Actual data')
plt.plot(x_test, y_pred, color='red', label='Predictions')
plt.title('Population Growth predictor')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```
